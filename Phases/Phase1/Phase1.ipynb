```python
emp_df = spark.read.csv("/Workspace/Users/iluqmanshaik255@gmail.com/Drafts/employees.csv" ,
                   header = True)
emp_df.createOrReplaceTempView("employees")
```

```python
%sql
--RETIREVE EMPLOYEES WHOSE AGE IS ABOVE 30
SELECT * FROM employees
WHERE age > 30;
```

```python
#FILTER EMPLOYEES WHOSE AGE IS ABOVE 30
emp_df.filter(emp_df.age > 30 ).show()
```

```python
emp_df.where("age > 30").show()
```

```python
%sql
--RETRIEVE EMPLOYEES IS ABOVE 30 AND SALARY EXCEEDS 50000
SELECT * FROM employees
WHERE age > 30 AND salary > 50000;
```

```python
#FILTER USING AND
emp_df.filter((emp_df.age > 30) & (emp_df.salary > 50000)).show()
```

```python
%sql
--RETRIEVE EMPLOYEES WHOSE ABOVE 30 OR BELONG TO HR
SELECT * FROM employees
    where AGE > 30 OR DEPARTMENT = "HR" ;
```

```python
#FILTER USING OR operator
emp_df.filter((emp_df.age > 30) | (emp_df.department == "HR")).show()
```

```python
%sql
-- RETRIEVE EMPLOYEES WHOSE NOT BELONG TO HR
SELECT * FROM employees
WHERE DEPARTMENT <> "HR";
```

```python
from pyspark.sql.functions import col

emp_df.filter(~(col("department") == "HR")).show()
```

```python
%sql 
--Retrieve employees from the HR and Finance 
SELECT * FROM employees 
WHERE department IN("HR"  , "Finance");
```

```python
# Filter rows using isin() 
emp_df.filter(emp_df.department.isin("HR" , "Finance")).show()
```

```python
%sql
-- Retrieve employees whose names start with 'A'
SELECT * FROM employees 
WHERE name LIKE "A%";
```

```python
# Filter names starting with the letter A
emp_df.filter(emp_df.name.startswith("A")).show()
```

```python
%sql
-- Retrieve employees whose names end with 'son'
SELECT * FROM employees
WHERE name LIKE "%n";
```

```python
# Filter names ending with "son"
emp_df.filter(emp_df.name.endswith("n")).show()
```

```python
#FILTER MATCHING USING LIKE()
emp_df.filter(emp_df.name.like("A%")).show()
```

```python
%sql
-- retrieve employees whose name contain 'son'
SELECT * FROM employees
WHERE name regexp "son";
```

```python
#perform regurlar expression
emp_df.filter(emp_df.name.rlike("son")).show()
```

```python
%sql
--   retrieve employees whose age is null
SELECT * FROM employees
WHERE age IS NULL;
```

```python
#FILTER ROWS WHERE AGE IS NULL
emp_df.filter(emp_df.age.isNull()).show()
```

```python
%sql
-- retrieve employees whose department is not null
SELECT * FROM employees 
WHERE department IS NOT NULL;
```

```python
#FILTER ROWS WHERE DEPARTMENT IS NOT NULL
emp_df.filter(emp_df.department.isNotNull()).show()
```

```python
%sql
-- RETRIEVE EMPLOYEES WITH A SALARY GREATER THAN 60000
SELECT * FROM employees 
WHERE salary > 60000;
```

```python
#FILTER EMPLOYEES WITH SALARY ABOVE 60000 
emp_df.filter(emp_df.salary > 60000).show()
```
